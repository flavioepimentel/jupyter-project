```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/countries.csv')
```

```python
data
```

```python
# Cheking the return
data.country
```

```python
# Getting only the true rows
br = data[data.country == 'Brazil']
br
```

```python
us = data[data.country == 'United States']
us
```

```python
ch = data[data.country == "China"]
ch
```

```python
# Comparing
plt.plot(br.year, us.population / 10 ** 6)
plt.plot(br.year, ch.population / 10 ** 6)
plt.plot(br.year, br.population / 10 ** 6)
plt.title('Population growth')
plt.xlabel('Years')
plt.ylabel('Population MM')
plt.legend(['United States', 'China', 'Brazil'])
plt.show()
```

```python
# Comparing the percentual value
plt.plot(br.year, us.population / us.population.iloc[0] * 100)
plt.plot(br.year, ch.population / ch.population.iloc[0] * 100)
plt.plot(br.year, br.population / br.population.iloc[0] * 100)
plt.title('Percentual growth')
plt.xlabel('Years')
plt.ylabel('% Growth')
plt.legend(['United States', 'China', 'Brazil'])
plt.show()
```

```python
# Check the percentual growth year by year
for n in br:
    plt.plot(br.year, br.population.iloc[n] / br.population.iloc[0])

```
